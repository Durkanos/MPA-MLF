## Exercise 1.3

Write a script that will convert an arbitrary integer to a binary number (the number will be represented as a string with only *0* and *1*). Avoid using the *bin()* function or any other pythons default functions

```python
def decimal_to_binary(n):
    binary_str = ""

    if n == 0:
        return "0"

    while n > 0:
        remainder = n % 2
        binary_str += str(remainder)
        n //= 2

    return binary_str[::-1]

#test
number = 156
binary_representation = decimal_to_binary(number)
print("Binary representation of", number, "is:", binary_representation)
```

## Exercise 1.4 - The Fibonacci Sequence:

The Fibonacci Sequence is a series of numbers. The next number is found by adding up the two numbers before it. The first two numbers are 0 and 1. For example, 0, 1, 1, 2, 3, 5, 8, 13. The next number in this series above is 8 + 13 = 21



Create a function that takes an integer as an input and returns a list that contains all Fibonacci numbers smaller than the input integer

example:

*print(fibonaci(10))*

[0, 1, 1, 2, 3, 5, 8]


```python
def fibonacci(max_number: int) -> list:
    fib_list = []
    a, b = 0, 1
    while a < max_number:
        fib_list.append(a)
        a, b = b, a + b
    return fib_list

#test
max_number = 72
result = fibonacci(max_number)
print(result)
```

```python
import random

def rock_paper_scissors() -> None:
    print("Welcome to Rock, Paper, Scissors!")
    user_choice = input("Enter your choice (Rock, Paper, or Scissors): ").lower()
    computer_choice = random.choice(["rock", "paper", "scissors"])

    print("You chose:", user_choice)
    print("Computer chose:", computer_choice)

    if user_choice == computer_choice:
        print("It's a tie!")
    elif (user_choice == "rock" and computer_choice == "scissors") or \
         (user_choice == "paper" and computer_choice == "rock") or \
         (user_choice == "scissors" and computer_choice == "paper"):
        print("You win!")
    else:
        print("You lose!")

#test
rock_paper_scissors()
```

```python
#Loops

def set_values_smaller_than_threshold_with_loop(arr, threshold):
    for i in range(arr.shape[0]):
        for j in range(arr.shape[1]):
            if arr[i, j] < threshold:
                arr[i, j] = 0
    return arr

#Test
result_with_loop = set_values_smaller_than_threshold_with_loop(arr.copy(), threshold)
print("Array after setting values smaller than", threshold, "to 0 (with loop):\n", result_with_loop)
```

```python
#Time without loop
start_time = time.time()
result_no_loop = set_values_smaller_than_threshold(arr.copy(), threshold)
end_time = time.time()
print("Execution time without loop:", end_time - start_time)

#Time with loop
start_time = time.time()
result_with_loop = set_values_smaller_than_threshold_with_loop(arr.copy(), threshold)
end_time = time.time()
print("Execution time with loop:", end_time - start_time)
```
